```python
df.loc[df['country'] == 'United States', :] # first is rows, second is columns. : is just clarifying
```

```python
df.loc[(df['country'] == 'United States') & (df['year']==1982), :] # first is rows, second is columns. : is just clarifying
```

```python
df.groupby(['year'])['lifeExp'].mean()  #calculaute mean by year for the life expentency column
```

```python
import numpy as np

df.groupby(['year'])['lifeExp'].agg(np.mean)  #calculaute mean by year for the life expentency column
```

```python
df.groupby(['year'])['lifeExp'].agg(np.std)
```

```python
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].agg(np.mean) #creates a hierarchical index
```

```python
df.groupby(['year','continent'])[['lifeExp','gdpPercap']].agg(np.mean).reset_index() #creates a regular dataframe
```
