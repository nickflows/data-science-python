```python
billboard = pd.read_csv('billboard.csv')
```

```python
billboard.head()
```

```python
billboard.melt(id_vars=['year','artist','track','time','date.entered'],
               value_name='rank',
               var_name='week') 

#tidy data - each column represents a variable, each row represents a single observation
```

```python
(
billboard
    .melt(id_vars=['year','artist','track','time','date.entered'],
         value_name='rank',
         var_name='week')
    .groupby('artist')['rank']
    .mean()
)

# . chaning everything
```

```python
billboard\
    .melt(id_vars=['year','artist','track','time','date.entered'],
         value_name='rank',
         var_name='week')\
    .groupby('artist')['rank']\
    .mean()

# . chaning everything
```

```python
(
billboard
    .melt(id_vars=['year','artist','track','time','date.entered'],
         value_name='rank',
         var_name='week')
    .groupby('artist')['rank']
    .mean()
)

# . chaning everything
```

```python
ebola = pd.read_csv('country_timeseries.csv')
```

```python
ebola.head()
```

```python
#create a tidyb dataset from the ebola dataset
ebola_long = ebola.melt(id_vars=['Date','Day'],
                       var_name= 'cd_country',
                       value_name='count') 
```

```python
ebola_long.head()
```

```python
#example string manipulation
# splits by underscore

'hello_world'.split('_')
```

```python
#split by string in pandas
# .dt --> datatime --> month
# .cat --> category 
ebola_long['cd_country'].str.split('_')
```

```python
# get data frame back
ebola_split = ebola_long['cd_country'].str.split('_', expand=True)
ebola_split
```

```python
ebola_long['test'] = 1
ebola_long.head()
```

```python
# add new columns from the split into the dataframe
ebola_long[['status','country']] = ebola_split
```

```python
# check results with new columns
print(ebola_long.head())
print(len(ebola_long))
```
