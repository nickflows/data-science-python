```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
import matplotlib.pyplot as plt
```

```python
tips.tip.plot(kind='hist', xlabel='count')
```

```python
cts = tips.smoker.value_counts()
```

```python
cts
```

```python
cts.plot(kind='bar')
```

```python
sns.countplot(x = 'smoker', data=tips)
```

```python
sns.displot(tips.total_bill, kde=False)
```

```python
#using seaborn to plot data
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', fit_reg=False)
```

```python
tips.columns
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips, hue='sex', fit_reg=False, col='smoker', row='day')
```

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter,'total_bill','tip')
```

```python
#gives back a figure object with axis

fig, (ax1, ax2) = plt.subplots (1, 2)

ax1.scatter(tips.tip, tips.total_bill)
ax2.hist(tips.total_bill)
```

```python
dis = sns.displot(tips.tip)
```

```python
#plot using seaborn and matplotlib
fig, (ax1, ax2) = plt.subplots(1,2)
sns.histplot(tips.tip, ax=ax1)
sns.regplot(x='total_bill', y='tip', data=tips, ax=ax2)
```
